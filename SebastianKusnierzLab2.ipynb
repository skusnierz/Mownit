```python
def is_zero(vector):
    #np.where(vector == 0) wyświetli tablicę indeksów dla których wartości tablicy vector będą równe 0
    print(np.where(vector == 0))
    #np.nonzero(vector) wyświetli tablicę indeksów dla których wartości tablicy vector będą różne od 0
    print(np.nonzero(vector))
    #np.argwhere działa podobnie jak np.where
    print(np.argwhere(vector == 0))


def is_inf(vector):
    #np.where(vector == np.inf) wyświetli tablicę indeksów dla których wartości tablicy vector będą równe inf
    print(np.where(vector == np.inf))
    #np.isinf(vector) wyświetli tablicę booli jeśli wartości odpowiedniej współrzędnek z vectora jest równa inf to w tablicu wynikowej pojawi się true jeśli nie to false 
    print(np.isinf(vector))
     #np.argwhere działa podobnie jak np.where
    print(np.argwhere(vector == np.inf))
    


def is_equal(array_a, array_b):
    print("First array :\n{}".format(array_a))
    print("Second array :   \n{}".format(array_b))
    print("Which cells are equals?  : \n{}".format(np.equal(array_a, array_b)))


matrix_a = np.matrix('1 2; 3 4; 2 3')
matrix_b = np.matrix('1 2; 2 4; 2 3')
matrix_c = np.matrix('1 2; 3 4; 2 3')
vector_a = np.array([1, 2, 3, 4, np.inf])
vector_b = np.array([1, 2, 3, 4, 0])
vector_c = np.array([2, 2, 4, 4, np.inf])

print("-------------Is Equal?--------------")
is_equal(matrix_a, matrix_b)
is_equal(vector_a, vector_b)
print("-----------------------------------")
print("-------------Is Zero?--------------")
is_zero(vector_b)
print("-----------------------------------")
print("-------------Is Inf?---------------")
is_inf(vector_a)
print("-----------------------------------")
```

### Zadanie 5
Napisz kilka programów, które demonstrują jak można w numpy tworzyć wektory i macierze wypełnione różnymi danymi lub mającymi pewną szczególną postać, np. , danymi losowymi, danymi z podanego zakresy, macierze identyczności, ... 

```python
#Tworzy wektor o podanym rozmiarze size oraz liczby będą losowane z zakresu [min, max]
def create_rand_vector(min, max, size):
    return np.random.randint(low=min, high=max+1, size=(size,))


#Tworzy macierz o podanym rozmiarze cols na rows oraz liczby będą losowane z zakresu [min, max]
def create_rand_matrix(min, max, cols, rows):
    return np.random.randint(low=min, high=max + 1, size=(rows, cols))


#Tworzy macierz identyczności o podanym rozmiarze size
def create_identity_matrix(size):
    return np.identity(size, dtype=None)


#Tworzy miacierz odwrotną do podanej
def create_inverse_matrix(matrix):
    return np.linalg.inv(matrix)


print(create_rand_vector(9, 10, 20))
print(create_rand_matrix(9, 15, 3, 5))
print(create_identity_matrix(5))
print(create_inverse_matrix(create_rand_matrix(3, 10, 3, 3)))
print(create_inverse_matrix(create_identity_matrix(5)))
```

### Zadania 6
Napisz programy wykonujący podstawowe operacje algebry: dodawanie wektórów, mnożenie wektorów, mnożenia macierzy. Najpierw napisz je w wersji szkolnej, nie zastanawiając się nad jakąkolwiek optymalizacją. Następnie spróbuj zmienić organizację obliczeń tak by uzyskać poprawę wydajności. Na koniec porównaj swoje metody do dedykowanych funkcji z bibliotek NumPy i SciPy. 


```python
import time
from tabulate import tabulate


def sum_of_vectors_simple_sol(vector_a, vector_b):
    start = time.time()
    result = []
    index = 0
    for value in vector_a:
        result.append(value + vector_b[index])
        index += 1
    end = time.time()
    return result, end - start


def product_of_vector_simple_sol(vector_a, vector_b):
    start = time.time()
    result = 0
    index = 0
    for value in vector_a:
        result += value * vector_b[index]
        index += 1
    end = time.time()
    return result, end-start


def product_of_matrix_simple_sol(matrix_a, matrix_b):
    start = time.time()
    num_rows_a, num_cols_a = matrix_a.shape
    num_rows_b, num_cols_b = matrix_b.shape
    result = np.zeros((num_rows_a, num_cols_b))
    for i in range(num_rows_a):
        for j in range(num_cols_b):
            for k in range(num_rows_b):
                result[i][j] += matrix_a[i][k] * matrix_b[k][j]
    end = time.time()
    return result, end-start


def sum_of_vectors_numpy(vector_a, vector_b):
    start = time.time()
    result = np.add(vector_a, vector_b)
    end = time.time()
    return result, end - start


def product_of_vector_numpy(vector_a, vector_b):
    start = time.time()
    result = np.dot(vector_a, vector_b)
    end = time.time()
    return result, end - start


def product_of_matrix_numpy(matrix_a, matrix_b):
    start = time.time()
    result = np.dot(matrix_a, matrix_b)
    end = time.time()
    return result, end-start


vector_a = np.array([2, 2, 4, 4, 2])
vector_b = np.array([1, 2, 3, 4, 0])
vector_c = np.array([1, 2])
vector_d = np.array([3, -5])
matrix_a = np.array([[12, 7, 3], [4, 5, 6], [7, 8, 9]])
matrix_b = np.array([[5, 8, 1, 2], [6, 7, 3, 0], [4, 5, 9, 1]])


output = [
        ['Sum of vectors (Simple solution) : ', sum_of_vectors_simple_sol(vector_a, vector_b)[0], sum_of_vectors_simple_sol(vector_a, vector_b)[1]],
        ['Sum of vectors (Numpy) : ', sum_of_vectors_numpy(vector_a, vector_b)[0], sum_of_vectors_numpy(vector_a, vector_b)[1]],
        ['Product of vectors (Simple solution) : ', product_of_vector_simple_sol(vector_a, vector_b)[0], product_of_vector_simple_sol(vector_a, vector_b)[1]],
        ['Product of vectors (Numpy) : ', product_of_vector_numpy(vector_a, vector_b)[0], product_of_vector_numpy(vector_a, vector_b)[1]],
        ['Product of matrix (Simple solution) : ',  product_of_matrix_simple_sol(matrix_a, matrix_b)[0], product_of_matrix_simple_sol(matrix_a, matrix_a)[1]],
        ['Product of matrix (Numpy) : ',  product_of_matrix_numpy(matrix_a, matrix_b)[0], product_of_matrix_numpy(matrix_a, matrix_a)[1]]
        ]

print(tabulate(output, headers=['L.P.', 'Value', 'Execute time '], tablefmt="fancy_grid"))
```
